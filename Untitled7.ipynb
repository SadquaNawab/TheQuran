```python
from colorama import Fore, Style
```

```python
import pyquran as q
```

```python
from googletrans import Translator, constants
from pprint import pprint
```

```python
translator = Translator()

def translate_arabic_to_english(text):
    translation = translator.translate(text, src="ar", dest="en")
    return translation.text
```

```python
def print_with_color(s, color=Fore.WHITE, brightness=Style.NORMAL, **kwargs):
    print(f"{brightness}{color}{s}{Style.RESET_ALL}", **kwargs)
```

```python
print_with_color("THE QURAN", color=Fore.GREEN)
```

```python
def func1():
    n = int(input("Enter surah Number: "))
    ans = q.quran.get_sura(n, with_tashkeel=False, basmalah=True)
    max_line_length = 200  # Maximum allowed length for a line
    for line in ans:
        if len(line) > max_line_length:
            # Split the line into smaller chunks
            chunks = [line[i:i+max_line_length] for i in range(0, len(line), max_line_length)]
            for chunk in chunks:
                print_with_color(chunk, color=Fore.GREEN)
                translated_chunk = translate_arabic_to_english(chunk)
                print_with_color(translated_chunk, color=Fore.BLUE)
        else:
            print_with_color(line, color=Fore.GREEN)
            translated_line = translate_arabic_to_english(line)
            print_with_color(translated_line, color=Fore.BLUE)
```

```python
def func2():
    s = int(input("Enter surah Number: "))
    l = int(input("Enter verse Number: "))
    ans = q.quran.get_verse(s, l, with_tashkeel=False)

    translated_ans = translate_arabic_to_english(ans)

    print_with_color(translated_ans, color=Fore.BLUE)


```

```python
def func3():
  n=int(input("Enter Surah Number:"))
  ans=q.quran.get_sura_name(n)
  print_with_color(ans, color=Fore.GREEN)
  translated_line = translate_arabic_to_english(ans)
  print_with_color(translated_line, color=Fore.BLUE)

```

```python
def main():

    while True:
        print_with_color("Choose From Below", color=Fore.BLUE)
        print_with_color("1-Get Sura", color=Fore.BLUE)
        print_with_color("2-Get Verse", color=Fore.BLUE)
        print_with_color("3-Get Sura Name", color=Fore.BLUE)
        n = int(input("Enter 1-3: "))
        if n < 1 or n > 4:
            print_with_color("Invalid input", color=Fore.RED)
            continue
        if n == 1:
            func1()
        elif n == 2:
            func2()
        elif n == 3:
            func3()
        ch=input("do you Want to continue (y/n)")
        if ch=='n':
          print_with_color("Allah Hafiz",Fore.GREEN)
          break


main()
```
